```python
import pandas as pd

df = pd.read_csv("student_management_messy.csv")

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert datatypes
df = df.copy()

df["CGPA"] = pd.to_numeric(df["CGPA"], errors="coerce")
df["Total_Marks"] = pd.to_numeric(df["Total_Marks"], errors="coerce")
```

```python
df["Registration_Date"] = pd.to_datetime(
    df["Registration_Date"],
    errors="coerce"
)

```

```python
# Numerical columns
num_cols = df.select_dtypes(include="number").columns
```

```python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Categorical columns
cat_cols = df.select_dtypes(include="object").columns
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Cleaning Completed")
```

```python
df
```

```python
df["Gender"] = df["Gender"].astype(str).str.strip().str.lower()

df["Gender"] = df["Gender"].replace({
    "m": "Male",
    "male": "Male",
    "f": "Female",
    "female": "Female",
    "other": "Other"
})
# Fill missing values
df["Gender"] = df["Gender"].replace("nan", pd.NA)
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df["Gender"].value_counts()
```

```python
# Department Column Cleaning
df["Department"] = df["Department"].astype(str).str.strip().str.upper()

df["Department"] = df["Department"].replace({
    "COMPUTER SCIENCE": "CSE",
    "COMPUTER SCIENCE ENGINEERING": "CSE",
    "AI&DS": "AI & DS",
    "AIDS": "AI & DS",
    "ARTIFICIAL INTELLIGENCE & DATA SCIENCE": "AI & DS",
    "ECE ": "ECE"
})
```

```python
#Missing Value fill
df["Department"] = df["Department"].replace("NAN", pd.NA)

df["Department"] = df["Department"].fillna(
    df["Department"].mode()[0]
)
```

```python
print(df["Department"].unique())
print(df["Department"].value_counts())
df["Department"] = df["Department"].astype(str).str.strip().str.upper()

df["Department"] = df["Department"].replace({
    "CS": "CSE",
    "COMP SCI": "CSE",
    "COMPUTER SCIENCE": "CSE",
    "CSE": "CSE"
})
```

```python
df["Placement_Status"] = (
    df["Placement_Status"]
    .astype(str)
    .str.strip()
    .str.upper()
)

df["Placement_Status"] = df["Placement_Status"].replace({
    "PLACED": "Placed",
    "UNPLACED": "Unplaced",
    "NOT ELIGIBLE": "Not Eligible"
})

df["Placement_Status"] = df["Placement_Status"].replace("NAN", pd.NA)

df["Placement_Status"] = df["Placement_Status"].fillna(
    df["Placement_Status"].mode()[0]
)
```
